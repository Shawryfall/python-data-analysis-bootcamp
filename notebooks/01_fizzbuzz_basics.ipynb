# FizzBuzz Basics in Python

This notebook demonstrates the classic FizzBuzz problem, often used in programming interviews to test basic control flow and modulo arithmetic.

## 1. Problem description

For each integer from 1 to 100:

- If the number is divisible by 3, print `fizz`
- If the number is divisible by 5, print `buzz`
- If the number is divisible by both 3 and 5, print `fizzbuzz`
- Otherwise, print the number itself

## 2. Simple loop implementation

```
for i in range(1, 101):
    if i % 15 == 0:
        print("fizzbuzz")
    elif i % 3 == 0:
        print("fizz")
    elif i % 5 == 0:
        print("buzz")
    else:
        print(i)
```

## 3. Wrap FizzBuzz in a function

Wrapping the logic in a function makes it reusable and easier to test.

```
def fizzbuzz(n: int) -> str:
    """Return 'fizz', 'buzz', 'fizzbuzz', or the number as a string."""
    if n % 15 == 0:
        return "fizzbuzz"
    elif n % 3 == 0:
        return "fizz"
    elif n % 5 == 0:
        return "buzz"
    else:
        return str(n)

# Quick check for the first 20 numbers
for i in range(1, 21):
    print(i, "->", fizzbuzz(i))
```

## 4. Generating a list of FizzBuzz outputs

Instead of printing, we can build a list. This is useful if we want to test or further process the results in code.

```
results = [fizzbuzz(i) for i in range(1, 101)]
results[:20]  # show the first 20 elements
```

In this notebook we:

- Encapsulated the logic in a reusable `fizzbuzz` function
- Generated a list of FizzBuzz outputs for further use
